```python
df = pd.read_csv("../input/englishsarcasm/SarcasmEnglish.csv")
df.head()
```

```python
len(df)
```

```python
from console_progressbar import ProgressBar
from googletrans import Translator

translator = Translator()
pb = ProgressBar(total = len(df),prefix='Here', suffix='Now', decimals = 2, length=150, fill='X', zfill='-')

for i in range(len(df)):
    df['Comments'][i] = translator.translate(df['Comments'][i], src='en', dest='bn').text
    #time.sleep(.001)
    pb.print_progress_bar(i)
    
    #if (i % 100 == 0):
        #print("Translated: {} Remaing: {}".format(i,len(df) - i))
```

```python
df.to_csv('TranslatedSarcasm.csv', index = False) 
```
